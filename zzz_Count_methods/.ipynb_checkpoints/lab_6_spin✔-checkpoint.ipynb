```python
def find_max(AAA):#find index of the max matrix el-t
    ind=[]
    compare=0
    for i in range(len(AAA)-1): #j
        for j in range(i+1,len(AAA)):#k
            if abs(AAA[i][j])>abs(compare):
                ind.clear()
                ind.append(i)
                ind.append(j)
                compare=AAA[i][j]
                #print(compare,ind)
                continue
            else:
                continue
                
    #print('max abs el-t: ',compare,ind)
    return ind
```

```python
def phi_funk(AAA,ind):#calculate the phi angle
    i=ind[0]
    j=ind[1] 
    
    if(AAA[i][i]-AAA[j][j])==0:
        phhhi=0.25*(math.pi)#1/2 * pi/2= 1/4 * pi
        #print('diffence is 0,angle phi is: ',phhhi)
    else:
        angle=(2*AAA[i][j])/(AAA[i][i]-AAA[j][j])
        phhhi=0.5*(math.atan(angle))
        #print(angle,'angle phi is: ',phhhi)
        
    return  phhhi
```

```python
def Jac(AAA,ind,phi):#Multiplication  of matrix
    JJ=[]
    JJ_t=[]
    AAA_finalle=[]
    AAA_finalle2=[]
    
    i_spez=ind[0]
    j_spez=ind[1] 
    #print(i_spez,j_spez)#idexes of main number in matrix
    for i in range(len(AAA)):
        JJ.append([])
        JJ_t.append([])
        AAA_finalle.append([])
        AAA_finalle2.append([])
            
    #print(JJ)
    
    for i in range(len(AAA)):
        for j in range(len(AAA)):
            
            if (i == i_spez and j == i_spez) or (i == j_spez and j == j_spez):
                JJ[i].append(round(math.cos(phi),10))#JJ[i].append(math.cos(phi))
                JJ_t[i].append(round(math.cos(phi),10))
                continue
            elif (i == i_spez and j == j_spez):
                JJ[i].append(round(-math.sin(phi),10))#JJ[i].append(-math.sin(phi))
                JJ_t[i].append(round(math.sin(phi),10))
                continue
            elif (i == j_spez and j == i_spez):
                JJ[i].append(round(math.sin(phi),10))#JJ[i].append(math.sin(phi))
                JJ_t[i].append(round(-math.sin(phi),10))
                continue
            elif j==i:
                JJ[i].append(1)
                JJ_t[i].append(1)
                continue
                
            else:
                JJ[i].append(0)
                JJ_t[i].append(0)
                
    #for i in range(3):
    #    print(JJ[i])
    #print()
    #for i in range(3):
    #    print(JJ_t[i])
    #print()
    
    #Matrix multiplication JJ_t*A:✔✔✔
    str_sum=0
    for strrr in range(len(AAA)):
        for i in range(len(AAA)):
            str_sum=0
            for j in range(len(AAA)):
                str_sum+=JJ_t[strrr][j]*AAA[j][i]#here we multiply str of JJ_t on column AAA 
                
            AAA_finalle[strrr].append(str_sum)
            
    #for i in range(3):
    #    print(AAA_finalle[i])
    #print()
    
    #Matrix multiplication AAA_nsw * JJ:✔✔✔??? something going wrong !!!
    str_sum=0
    for strrr in range(len(AAA)):
        for i in range(len(AAA)):
            str_sum=0
            for j in range(len(AAA)):
                str_sum+=AAA_finalle[strrr][j]*JJ[j][i]#here we multiply str of AAA_nsw on column JJ 
                
            AAA_finalle2[strrr].append(str_sum)
            
    #for i in range(3):
    #    print(AAA_finalle2[i])
    #print()
    
    return AAA_finalle2
```

```python
def checkky(AAA):#here we check our "win" condition
    summa=0
    for i in range(len(AAA)):
        for j in range(len(AAA)):
            if i == j:
                continue
            else:
                summa+=(AAA[i][j])**2
    
    
    return (summa**0.5)
```

```python
import copy

def solve_lin_equ1(A,B):
#Дано:А-матрица СЛАУ, В-вектор-столбец свободных членов СЛАУ
#Рез.: Если detA~=0, то х-вектор-столбец решение СЛАУ, в противном случае, если система совместна, то x(:,1)-частное решение, x(:,2:end)  
#содержит в столбцах базис решения, соответствующий СЛАУ, либо x=[], если СЛАУ несовместная
    AAA=copy.deepcopy(A)
    for i in range(len(A)):#razshirennaia matrix
        AAA[i].append(B[i])
        print(AAA[i])
        
    print()
    column_massive=[]#слобцы нашей марицы как список
    for i in range(len(A)):
        column_massive.append([])
        for j in range(len(A)):
                column_massive[i].append(A[j][i])
        #print(column_massive[i])
        
    #прямой ход:
    for i in range(len(A)):
        #print(i)
        
        if i == len(A)-1:
            break
            
        else:
            max_ind=max(column_massive[i][i:],key=abs)#here we just get current number - not index
            #print('max abs: ',max_ind)
            if abs(max_ind) < 0.000001: #if in column all zero - sys is ... have 00 roots
                continue

            max_ind=column_massive[i].index(max_ind)#index of str there is max abs el-t
            #print('max abs index: ',max_ind)
            ##########up str (here we change str`s): put it on diagonalle
            box1=[]
            box1.clear()
            box1=copy.deepcopy(AAA[i])

            AAA[i].clear()
            AAA[i]=(AAA[max_ind])#copy.deepcopy
            AAA[max_ind]=copy.deepcopy(box1)

            #for i1000 in range(len(A)):#razshirennaia matrix
            #    print(AAA[i1000])
            #######################################

            #differece between column el-ts:
            for i2 in range(i,len(A)-1):#str
                under_main=((-AAA[i2+1][i])/AAA[i][i])
                #under_main=(AAA[i][i]/AAA[i][i])*(-AAA[i2][i])#koef under main el-t
                for j2 in range(i,len(A)):#col
                    AAA[i2+1][j2]=AAA[i2+1][j2] + under_main*AAA[i][j2]
                    if abs(AAA[i2+1][j2])<0.000001:
                        AAA[i2+1][j2]=0

            for i2 in range(len(A)):
                column_massive[i2].clear()
                for j2 in range(len(A)):
                    boxxx=0
                    boxxx=copy.deepcopy(AAA[j2][i2])
                    column_massive[i2].append(boxxx)

            #print('after:')
            #for i in range(len(A)):#razshirennaia matrix
            #    print(AAA[i])

    #print('transform:')
    #for i in range(len(A)):
    #    print(column_massive[i])
    
    #print('\nreverse step:\n')
#####################################################################################################################
#####################################################################################################################
    #обратный ход:
    i_spez=len(A)
    i_spez2=len(A)-1
    for i in range(len(A)):
        
        i_spez-=1
        #print('hohoho',i_spez,'i:',i)
        
        if i_spez < 1:
            break
            
        else:
        
            max_ind=AAA[i_spez][i_spez]    
            if abs(max_ind) < 0.000001: #if in column all zero - sys is ... have 00 roots
                continue

            #print('max abs index: ',i_spez)

            #######################################

            #differece between column el-ts:
            
            for i2 in range(i_spez+1):#str
                if i2==0:
                    continue
                
                upper_main=(-AAA[i_spez-i2][i_spez])/AAA[i_spez][i_spez]
                
                for j2 in range(i+2):#col
                    #print(j2,AAA[i_spez-i2][-j2-1])
                    AAA[i_spez-i2][-j2-1]= AAA[i_spez-i2][-j2-1] + upper_main*AAA[i_spez][-j2-1]
                    
                    if abs(AAA[i_spez-1][i_spez])<0.000001:
                        AAA[i_spez-1][i_spez]=0
                        
            print('after:')
            for i in range(len(A)):#razshirennaia matrix
                print(AAA[i])
    
    print('\nlast part:\n')
    for i in range(len(A)):
        devider=AAA[i][i]
        if abs(devider) <0.000001:
            print(AAA[i])
            continue
        for j in range(len(A)):
            AAA[i][j]=AAA[i][j]/devider
        print(AAA[i])
    
```
