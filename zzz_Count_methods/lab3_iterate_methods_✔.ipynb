```python
#method of simple iterations
#check max diagonal el-ts

Tmp.clear()
X.clear()

delta=10000
stop=0
devvi=[]

for j in range(len(A[1])):
    Tmp.append(0)
    X.append(0)

while delta >= eps:
    devvi.clear()
    stop+=1
    if stop==100:
        break
        
    for i in range(len(A[1])):
        box=0
        box+=B[i]
        ##X[i]+=B[i]
        for j in range(len(A[1])):           
            if j==i:
                continue
            box+=(-Tmp[j])*A[i][j]
            ##X[i]+=(-Tmp[j])*A[i][j]

        X[i]=box/A[i][i]
        ##X[i]=X[i]/A[i][i]
        #print(X)
    
    for i in range(len(A[1])):
        devvi.append(X[i]-Tmp[i])
    delta=round(abs(max(devvi,key=abs)),6)
    #print('max difference: ',delta)
    
    for i in range(len(A[1])):
        Tmp[i]=X[i]
        
#print('last delta=',delta,' while epsilon is: ',eps)
print('Answ by simple iterations: ',X,stop)
    
```
