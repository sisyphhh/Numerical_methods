```python
def progonka(A_matr, B_matr):
    #n = int(input())
    n=len(B_matr)#заменить на размерность системы len(B_matr)
    
    mass2 = []
    Xi=[0 for i in range(n)]
    PQ=[]

    for i in range(len(A_matr)):
        mass2.append(list(np.append(A_matr[i],B_matr[i])))
        #print(mass2[i])
        
    #mass2 =[[-11, 9, 0, 0, 0, -158],[0, -8, -6, 0, 0, 66],[0, 6, 15, -2, 0, -45],[0, 0, 4, 6, -1, 24],[0, 0, 0, -7, -10, -1]]
    indx1=0
    for i in range(n):#first 'go'
      #print(PQ)
        if i == 0:
            p=(-mass2[0][1])/mass2[0][0]
            q=(mass2[0][n])/mass2[0][0]
            PQ.append([p,q])

        elif 1<=i<n-1:
            p=(-mass2[i][i+1])/(mass2[i][indx1] * PQ[i-1][0] + mass2[i][indx1+1])
            q=(mass2[i][n]-mass2[i][indx1]*PQ[i-1][1])/(mass2[i][indx1] * PQ[i-1][0] + mass2[i][indx1+1]) 
            #PiQi[0]=p
            #PiQi[1]=q
            PQ.append([p,q])

            indx1+=1
        else:#last line
            p=0
            q=(mass2[i][n]-mass2[i][indx1]*PQ[i-1][1])/(mass2[i][indx1] * PQ[i-1][0] + mass2[i][indx1+1]) 
            #PiQi[0]=p
            #PiQi[1]=q
            PQ.append([p,q])
      #print(PQ)
    #print('\n')   
    #print('Pi and Qi:',PQ)

    i=n-1
    while i >= 0:
        if i==n-1:
            Xi[i]=round(PQ[i][1],6)
            i-=1
        else:
            Xi[i]=round(PQ[i][0]*Xi[i+1]+PQ[i][1],6)
            i-=1
    #print('\n')
    #print('Xi:',Xi)
    '''masdiag=[]
    for i in range(n):
        masdiag.append(mass2[i][i])
    #print('diagonal:',masdiag)
    summa=0
    for i in range(n):
        summa+=abs(masdiag[i])'''
    #for i in range(n):
      #print(masdiag[i]/(summa-abs(masdiag[i])))
    '''dl=[]
    for i in range(n):
        if abs(masdiag[i])>= (summa - abs(masdiag[i])):
            dl.append(masdiag[i])
    dl.sort()    
    print(dl)'''
    return Xi
```

```python
def F_05(U,i1,i2,h1,h2,a=1,tau=1):#add j LAYER
    return 0.5*a*tau/(h2**2) * (U[i1][i2-1]+U[i1][i2+1]) + (1-a*tau/(h2**2))*U[i1][i2]
    
def F_1(U,i1,i2,h1,h2,a=1,tau=1):#add j LAYER
    return 0.5*a*tau/(h1**2) * (U[i1-1][i2]+U[i1+1][i2]) + (1-a*tau/(h1**2))*U[i1][i2]
    
def matr_maker(U,x,y,i11,i22,h1,h2,time,flag='x',tau=1,a=1):
    #print(tau)
    N_x=len(x)
    N_y=len(y)
    
    if flag=='x':#пробегаем по иксам 0.5 tau*(j+1/2))
        U_05=np.zeros((N_x-2,N_x-2))
        B_05=np.zeros(N_x-2)
        index=1;print('fix y')
        #for i1 in range(1,N_x-1):
               
        for i1 in range(N_x-2):
            if i1==0:
                U_05[i1][0]=-(1+a*tau/(h1**2))
                U_05[i1][1]=0.5*a*tau/(h1**2)
                B_05[i1]=-F_05(U,i1,i22,h1,h2,a=1,tau=tau) - 0.5*a*tau/(h1**2)*math.sinh(i22*h2)*math.exp(-3*a*(time+tau*1/2))#math.exp(-3*a*(0.5*tau+time))
            elif i1==N_x-3:
                U_05[i1][N_x-4]=0.5*a*tau/(h1**2)
                U_05[i1][N_x-3]=-(1+a*tau/(h1**2)-0.5*a*tau/(h1**2))
                B_05[i1]=-F_05(U,i1,i22,h1,h2,a=1,tau=tau) + h1*2*math.sinh(i22*h2)*math.exp(-3*a*(time+tau*1/2))#math.exp(-3*a*(0.5*tau+time))
            else:
                U_05[i1][index-1]=0.5*a*tau/(h1**2)
                U_05[i1][index]=-(1+a*tau/(h1**2))
                U_05[i1][index+1]=0.5*a*tau/(h1**2)
                B_05[i1]=-F_05(U,i1,i22,h1,h2,a=1,tau=tau)
                index+=1

        return [U_05,B_05]

    else:#пробегаем по игрекам 1.0 tau*(j)
        U_1=np.zeros((N_y-2,N_y-2))
        B_1=np.zeros(N_y-2)
        index=1;print('fix x')

        for i2 in range(N_y-2):
            if i2==0:
                U_1[i2][0]=-(+1+a*tau/(h2**2))+0.5*a*tau/(h2**2)
                U_1[i2][1]=0.5*a*tau/(h2**2)
                B_1[i2]=-F_1(U,i11,i2,h1,h2,a=1,tau=tau) - 0.5*a*tau/(h2)*math.cos(2*h1*i11)*math.exp(-3*a*(time+tau))#math.exp(-3*a*(tau+time))
            elif i2==N_y-3:
                U_1[i2][N_y-4]=0.5*a*tau/(h2**2)
                U_1[i2][N_y-3]=-(1+a*tau/(h2**2))
                B_1[i2]=-F_1(U,i11,i2,h1,h2,a=1,tau=tau) - 0.5*a*tau/(h2**2)*3/4*math.cos(2*h1*i11)*math.exp(-3*a*(time+tau))#math.exp(-3*a*(tau+time))
            else:
                U_1[i2][index-1]=0.5*a*tau/(h2**2)
                U_1[i2][index]=-(1+a*tau/(h2**2))
                U_1[i2][index+1]=0.5*a*tau/(h2**2)
                B_1[i2]=-F_1(U,i11,i2,h1,h2,a=1,tau=tau)
                index+=1

        return [U_1,B_1]
```

```python
st_X=0;end_X=round(math.pi/4,6)#;print(end_X)
st_Y=0;end_Y=round(math.log(2),6)#;print(end_Y)

N=11
eps=0.01

x,y = np.linspace(st_X,end_X,N),np.linspace(st_Y,end_Y,N)
print('x:',x,'\ny:',y,'\n')
#print(x,math.pi/4)
h_x=x[1]-x[0]
h_y=y[1]-y[0]


U_xy=np.zeros((N,N))
'''заполнение сетки начальными условиями'''
for i1 in range(N):
    for i2 in range(N):  
        U_xy[i1][i2]=math.cos(2*x[i1])*math.sinh(y[i2])
        #
    #print(U_xy[i1])

'''--------------------------------------------------------------------------'''
'''--------------------------------------------------------------------------'''
NNN=4
T=0.15#0.15#2
t=np.linspace(0,T,NNN);print("t layers ",t,'\n')
tau=round(t[1]-t[0],6);print('tau:',tau,'\n')
#j=0
a=1

U_xy_05=U_xy.copy()#np.zeros((N,N))
U_xy_1=U_xy.copy()#np.zeros((N,N))

collection_list=[]
collection_list.append(np.round(U_xy,6))

for j in range(0,len(t)-1):
    
    #U_xy_05=U_xy_1.copy()
    '''first step'''
    
    for i2 in range(1,len(y)-1):
        tmp=matr_maker(U_xy_1,x,y,i11='x',i22=i2,h1=h_x,h2=h_y,time=t[j],flag='x',tau=tau,a=1)#time=t[j]

        aa=progonka(A_matr=tmp[0], B_matr=tmp[1])
        
        #print('\n0.5_progon',aa,type(aa),len(aa))
        #print('gauss',np.linalg.solve(tmp[0], tmp[1]))

        for k in range(1,len(x)-1):
            U_xy_05[k][i2]=round(aa[k-1],6)
            
        U_xy_05[0][i2]=round(math.sinh(i2*h_y)*math.exp(-3*a*(t[j]+tau*1/2)),6)#math.exp(-3*a*(0.5*tau+t[j]))
        U_xy_05[-1][i2]=round(aa[-1]-h_x*2*math.sinh(i2*h_y)*math.exp(-3*a*(t[j]+tau*1/2)),6)#math.exp(-3*a*(0.5*tau+t[j]))#U_xy_05[-2][i2]
    
    for i in range(len(x)):
        U_xy_05[i][-1]=round(3/4*math.cos(2*x[i])*math.exp(-3*a*(t[j]+tau*1/2)),6)
    for i in range(len(x)):
        U_xy_05[i][0]=round(U_xy_05[i][1]-h_y*math.cos(2*x[i])*math.exp(-3*a*(t[j]+tau*1/2)),6)
    '''переопределяем граничные условия после пер-да на j+0.5'''
    #U_xy_05=granichnie(U_xy_05,x,y,h_x,h_y,t=t[j]+0.5*tau,tau=tau)
    
    print('\n=============================\n')
    
    #U_xy_1=U_xy_05.copy()
    '''second step'''
    for i1 in range(1,len(x)-1):
        tmp=matr_maker(U_xy_05,x,y,i11=i1,i22='y',h1=h_x,h2=h_y,time=t[j],flag='y',tau=tau,a=1)#time=t[j]

        aa=progonka(A_matr=tmp[0], B_matr=tmp[1])
        
        #print('\n1_progon',aa,type(aa),len(aa))
        #print('gauss',np.linalg.solve(tmp[0], tmp[1]))

        for k in range(1,len(y)-1):
            U_xy_1[i1][k]=round(aa[k-1],6)#U_xy_1[k+1][i2]=aa[k]
        U_xy_1[i1][0]=round(aa[0]-h_y*math.cos(2*x[i1])*math.exp(-3*a*(t[j]+tau)),6)#math.exp(-3*a*(t[j]+1))
        U_xy_1[i1][-1]=round(3/4*math.cos(2*x[i1])*math.exp(-3*a*(t[j]+tau)),6)#math.exp(-3*a*(t[j]+1))
    
    for i in range(len(y)):
        U_xy_1[0][i]=round(math.sinh(y[i])*math.exp(-3*a*(t[j]+tau)),6)
    for i in range(len(y)):
        U_xy_1[-1][i]=round(U_xy_1[-2][i]-h_x*2*math.sinh(y[i])*math.exp(-3*a*(t[j]+tau)),6)#U_xy_05[-2][i2]
    '''переопределяем граничные условия после пер-да на j+1'''
    #U_xy_1=granichnie(U_xy_1,x,y,h_x,h_y,t=t[j]+tau,tau=tau)
    
    testt=0
    testt=copy.deepcopy(U_xy_1)
    collection_list.append(testt) 
        
print(len(collection_list))

for i in range(len(collection_list)):
    with open ('table_4_%s.txt'%(i),'w') as outfile:
        for j in range(len(collection_list[i])):
            for k in range(len(collection_list[i][j])):
                outfile.write(str(collection_list[i][j][k])+';')
            outfile.write('\n')

'''        
with open (target+'table_last_%s_%s.txt'%(i1,dop),'a') as outfile:
            outfile.write(str(n)+'    '+str(round(t_i[-2][0],5))+'    '+str(start_i+1)+'    '+str(round(abs(t_i[-1][0]),5))+'    '+str(Delta_n)+'\n')
'''
```
