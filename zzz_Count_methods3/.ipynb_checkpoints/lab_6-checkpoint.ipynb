```python
def F(x):
    if x==0:
        return 1
    else:
        return round(1+1/x*(math.cos(x/2)-1),6)

def An_j(a,b,n):#центральные прямоугольники
    return round((b-a)/n,6)
     
def K_x_s(x_i,s_j):
    return round(math.sin(x_i)*(s_j),6)
```

```python
enddd=1/2

st_X=0;end_X=round(enddd,6)#;print(end_X)
st_T=0;end_T=round(enddd,6)#;print(end_Y)

N=11
#eps=0.01
liambda=1

x,t = np.linspace(st_X,end_X,N),np.linspace(st_T,end_T,N)
print('x:',x,'\nt:',t,'\n')
#print(x,math.pi/4)
h_x=x[1]-x[0]
h_t=t[1]-t[0]


U_n=np.zeros((N,N))
E_n=np.zeros((N,N))
F_n=np.zeros(N)
#print(F_n)
'''заполнение сетки начальными условиями'''
h=(end_X-st_X)/N
#h=(end_X-st_X)/(N-1)
for i in range(1,N+1):    
    x_i=st_X+(i-0.5)*h
    for j in range(1,N+1):  
        x_j=st_X+(j-0.5)*h
        U_n[i-1][j-1]= An_j(a=st_X,b=end_X,n=N)*K_x_s(x_i=x_i,s_j=x_j)
    
    if i==0:
        F_n[i-1]=1
    else:
        F_n[i-1]=F(x_i)
    
    E_n[i-1][i-1]=1
    

Answ=E_n-U_n
ans=np.round(np.linalg.solve(Answ, F_n),6)
print('U_i=',ans)
#print('gauss',np.linalg.solve(tmp[0], tmp[1]))
#np.sin(x)
```
