```python
import math
import numpy as np
hX=0.1
hY=0.1
tao=0.2

stX=0
edX=1
X=int((edX-stX)/hX+1)
stY=0
edY=1
Y=int((edY-stY)/hY)+1
stT=0
edT=1
T=int((edT-stT)/tao*2)+1

x=[round(stX+i*hX,2) for i in range(X)]
y=[round(stY+i*hY,2) for i in range(Y)]
t=[round(stY+i*tao,2) for i in range(T)]
X,Y,T
```

```python
# начальные и краевые условия
res=np.zeros((T,Y,X))
for i in range(X):
    for j in range(Y):
        res[0,j,i]=x[i]*y[j]
```

```python
for k in range(1,T): #,T
    if k%2!=0:
        # прогонка по x - получу внутренность нового слоя
        res[k]=Filling_by_Direction(res[k-1],tao,x,y,t[k])
    else:
        res[k]=Filling_by_Direction(res[k-1].T,tao,y,x,t[k])
        
```

```python
def Filling_by_Direction(res_prev, tao,x,y,tt):  
    #create new layer from i=1 to the end] i.e. len outer=Y-1 len inner = Y-2
    #print(res_prev)
    Y=len(res_prev)
    X=len(res_prev[0])
    hX=x[1]-x[0]
    hY=y[1]-y[0]
    new=np.zeros((Y,X)) #new layer +
    #print(new.shape,hX,hY) # (11,11)
    for j in range(1,Y-1):
        m=matrix(res_prev, tao,hX,hY,x,tt,y[j],j)
        #print(len(m),len(m[0]))
        X_line=gauss(m)
        #print(len(X_line)) #10
        #print(type(X_line))
        #print(X_line,'\n')
        new[j]=[0]+X_line.copy()
    for i in range(1,Y):
        new[-1,i]=new[-2,i]/(1-hY)
    return new
```

```python
#необазательно
def Filling_by_Y_Direction(res_prev, tao,x,y,tt):  
    Y=len(res_prev)
    X=len(res_prev[0])
    hX=x[1]-x[0]
    hY=y[1]-y[0]
    new=np.zeros((X,Y))
    for i in range(1,X-1):
        m=matrix(res_prev.T, tao,hY,hX,y,tt,x[i],i)
        X_line=gauss(m)
        new[i]=[0]+X_line.copy()
    for j in range(1,X):
        new[-1,j]=new[-2,j]/(1-hX)
    return new.T
```

```python
def matrix(res_prev, tao,hX,hY,x,tt,yy,j): #create and fill matrix fot fixed y[j]=y on tt layer
    X=len(res_prev[0])-1 #10
    
    mm=[[0 for _ in range(X+1)] for q in range(X)]
    
    #print(len(mm), len(mm[0])) #(10,11)
    mm[0][0]=(1/tao+2/hX/hX)
    mm[0][1]=-1/hX/hX
    mm[-1][-2]=1-1/hX
    mm[-1][-3]=1/hX

    for jj in range(1,X-1):  #filling diagonals
        mm[jj][jj]=(1/tao+2/hX/hX)
        mm[jj][jj-1]=-1/hX/hX
        mm[jj][jj+1]=-1/hX/hX
        
    for jj in range(X-1): #b-column
        mm[jj][-1]=( res_prev[j+1,jj+1]-2*res_prev[j,jj+1]+res_prev[j-1,jj+1])/hY/hY+ res_prev[j,jj+1]/tao +x[jj+1]*yy*math.sin(tt)
    return mm
```

```python
# Дано: Расширенная матрица A=(C|b), n x (n+1)
# Результат: решение системы Cx=b
def gauss(A):
    n = len(A) #размерность матрицы C

    for i in range(0, n):
        # Search for maximum in this column
        maxEl = abs(A[i][i])
        maxRow = i
        for k in range(i+1, n):
            if abs(A[k][i]) > maxEl:
                maxEl = abs(A[k][i])
                maxRow = k
        if (maxRow!=i):
            # Swap maximum row with current row (column by column)
            for k in range(i, n+1):
                tmp = A[maxRow][k]
                A[maxRow][k] = A[i][k]
                A[i][k] = tmp

        # Make all rows below this one 0 in current column
        for k in range(i+1, n):
            c = -A[k][i]/A[i][i]
            for j in range(i, n+1):
                if i == j:
                    A[k][j] = 0
                else:
                    A[k][j] += c * A[i][j] 
        

    # Solve equation Ax=b for an upper triangular matrix A
    x = [0 for i in range(n)]
    for i in range(n-1, -1, -1):
        x[i] = A[i][n]/A[i][i]
        for k in range(i-1, -1, -1):
            A[k][n] -= A[k][i] * x[i]
    return x


A=[[-2,-9,-3,7,-26],[-7,2,2,5,-25],[-6,2,0,0,-16],[0,-3,8,-3,-5]]
gauss(A)
```
