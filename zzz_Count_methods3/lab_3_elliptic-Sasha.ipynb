### Условия:


$$\frac{\partial^2 u}{\partial x^2} + \frac{\partial^2 u}{\partial y^2} = -2\frac{\partial u}{\partial x} - 2\frac{\partial u}{\partial y} - 4u$$

$$u(0,y) = e^{-y}\cos y, u(x,0) = e^{-x}\cos x$$

$$u(\frac{\pi}{2},y) = 0,u(x,\frac{\pi}{2}) = 0 $$

$$=====================================================================$$

$$u^{(k+1)}_{i,j} = \frac{1}{4(1-h^2)}[u^{(k)}_{i+1,j}+u^{(k)}_{i-1,j}+u^{(k)}_{i,j-1}+u^{(k)}_{i,j+1} + h\cdot2(u^{(k)}_{i+1,j} - u^{(k)}_{i-1,j} + u^{(k)}_{i,j+1} - u^{(k)}_{i,j-1}) ] $$

$$u_{i,j}^{(k+1)} = \frac{1}{(4-\frac{4}{h^2})}[-(\frac{1}{h^2}+\frac{1}{h})(u_{i+1,j}^{(k)}+u_{i,j+1}^{(k)}) + (\frac{1}{h} - \frac{1}{h^2})(u_{i-1,j}^{(k)}+u_{i,j-1}^{(k)})]$$

$$u^{(k)}_{0,j} = e^{-y}\cos y; u^{(k)}_{i,0} = e^{-x}\cos x$$

$$u^{(k)}_{N,j} = 0;u^{(k)}_{i,N} = 0 $$

```python
import numpy as np
import math

def linear_interpolation(x,y,res,h):#h=h_x
    Y,X=res.shape
    for j in range(1,Y-1):
        for i in range(1,X-1):     
            res[j][i]=res[j][0]+(res[j][-1]-res[j][0])*i*h/(math.pi/2)
    return res

def Liebmann_iter(res,h=0,h1=1,h2=1):#h for case h_x==h_y
    Y,X=res.shape
    res_new=res.copy()
    print('h=',h,'h1=',h1,' h2=',h2)
    for j in range(1,Y-1):
        for i in range(1,X-1):
            if h!=0:
                res_new[j][i]=(res[j][i+1]*(1+2*h*h) + res[j][i-1]*(1-2*h*h) + res[j+1][i]*(1+2*h*h) + res[j-1][i]*(1-2*h*h))/(4-4*h**2)
                #res_new[j][i]=((h**2)*res[j][i] + res[j][i+1] + res[j][i-1] + res[j+1][i] + res[j-1][i])/(4)
            else:
                #print('else')
                '''variant 1'''
                res_new[j][i]=(res[j][i+1]*(h2*h2+2*h1*h2*h2) + res[j][i-1]*(h2*h2-2*h1*h2*h2) + res[j+1][i]*(h1*h1+2*h1*h1*h2) + res[j-1][i]*(h1*h1-2*h1*h1*h2))/(2*h2*h2+2*h1*h1-4*h1*h1*h2*h2)
                
    return res_new

def norm(res):
    #print(res[5])
    Y,X=res.shape
    tmp=[0 for i in range(Y)]
    for j in range(Y):
            tmp[j]=max(res[j],key=abs) 
    print('\n',abs(max(tmp,key=abs)))
    return abs(max(tmp,key=abs))
```

```python
import copy
st_X=0;end_X=round(math.pi/2,6)#;print(end_X)
st_Y=0;end_Y=round(math.pi/2,6)#;print(end_Y)

N=11
eps=0.001

x,y = np.linspace(st_X,end_X,N),np.linspace(st_Y,end_Y,N)
print('x:',x,'\ny:',y,'\n')
#print(x,math.pi/4)
h_x=x[1]-x[0];print(h_x)
h_y=y[1]-y[0];print(h_y)


U_xy=np.zeros((len(y),len(x)))

'''заполнение сетки начальными условиями по U(0,y) & U(pi/2,y) & U(x,0) & U(x,pi/2)'''

for j in range(len(y)):#U(0,y)
    U_xy[j][0]=np.exp(-y[j])*np.cos(y[j])    

for j in range(len(y)):#U(pi/2,y)=0
    U_xy[j][-1]=0
    
for j in range(len(x)):#U(x,0)
    U_xy[0][j]=np.exp(-x[j])*np.cos(x[j])

for j in range(len(x)):#U(x,pi/2)=0
    U_xy[-1][j]=0       
    

'''вывод нашей краевой/граничной сетки'''    
for i in range(len(U_xy)):
    print(U_xy[i])
    

'''вычисление нулевой сетки - линейная интерполяция - первичное заполнение сетки(внутри)'''
res_old=linear_interpolation(x,y,U_xy,h=h_x).copy()

print('\n\nafter lin interpol1\n')
for i in range(len(U_xy)):
    print(U_xy[i])

    
'''заполнение внутренностей сетки по Методу Либмана'''
res_new=Liebmann_iter(res_old,h1=h_x,h2=h_y) # res_new=  вычисление первой итерация - по формуле
for i in range(len(res_new)):
    print(res_new[i],'\n')

k=1
#print('n',norm(res_new-res_old))
while norm(res_new-res_old)>eps and k<=50:
    
    res_old=res_new.copy()
    
    res_new=Liebmann_iter(res_old,h1=h_x,h2=h_y) # res_new=вычисление второй итерации - по формуле
    k+=1
    #print(norm(res_new-res_old),eps,k)# k, norm(res_new-res_old)
    print(k)

    
print('\nfinal table\n')
for i in range(len(res_new)):
    print(res_new[i])
```
