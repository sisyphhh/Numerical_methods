```python
import numpy as np
import matplotlib.pyplot as plt
from math import  *
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(111)
y = lambda a,x: np.sqrt(a**2 - x**2 )
size = 8
c = [0.3,0.3,0.3]
for a in range(1,21):
    size /= 1.05
    x = np.linspace(-a,a,5500)
    ax.plot(x,y(a,x), color = c, linewidth = size)
    ax.plot(x,-y(a,x), color = c, linewidth = size)
ax.set_xticks([])
ax.set_yticks([])
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_color('none')
ax.spines['left'].set_color('none')
plt.savefig('/Users/haru4me/Downloads/circle.jpeg',dpi = 200)
plt.show()
```

```python
fig = plt.figure(figsize = (10,10))
ax = fig.add_subplot(111)
M = 400
cx = M/2
cy = M/2
pic = np.zeros((M,M,3))
s = []

for i in range(pic.shape[0]):
    for j in range(pic.shape[1]):
        r = sqrt((i-cx)**2+(j-cy)**2)
        r /= (3-r/200)
        S = int(127*sin(r)+1)
        pic[i][j] = [S,S,S]
ax.set_xticks([])
ax.set_yticks([])
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_color('none')
ax.spines['left'].set_color('none')
plt.imshow(pic)
plt.savefig('/Users/haru4me/Downloads/circle4.jpeg',dpi = 200)
```
