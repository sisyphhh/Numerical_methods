### Условия:

$$\frac{\partial^2 u}{\partial x^2} + \frac{\partial^2 u}{\partial y^2} = -u$$

$$u(0,y) = 0$$
$$u_y(x,0) = sin{x}$$

$$u(\frac{\pi}{2},y) =-sin{y}$$
$$u_y(x,1)-u(x,1) = 0 - sin{y}----wtf???$$
$$=====================================================================$$

$$u^{(k+1)}_{i,j} = \frac{1}{(4-h^2)}[u^{(k)}_{i+1,j}+u^{(k)}_{i-1,j}+u^{(k)}_{i,j-1}+u^{(k)}_{i,j+1} ] $$


$$u^{(k)}_{0,j} =0$$
$$u^{(k)}_{i,0} = u_{i,1}-h*sin{x}$$ 

$$u^{(k)}_{N,j} = -sin{y}$$
$$u^{(k)}_{i,N} =  \frac{u_{i,N-1}}{(1-h)} $$

```python
import numpy as np
import math

def linear_interpolation(x,y,res,h):#h=h_x
    Y,X=res.shape
    for j in range(1,Y-1):
        for i in range(1,X-1):     
            res[j][i]=res[j][0]+(res[j][-1]-res[j][0])*i*h/(math.pi/2)
    return res

def Liebmann_iter(res,h=0,h1=1,h2=1):#h for case h_x==h_y
    Y,X=res.shape
    res_new=res.copy()
    print('h=',h,'h1=',h1,' h2=',h2)
    for j in range(1,Y-1):
        for i in range(1,X-1):
            if h!=0:
                res_new[j][i]=(res[j][i+1] + res[j][i-1] + res[j+1][i] + res[j-1][i])/(4-h**2)
                
                #res_new[j][i]=((h**2)*res[j][i] + res[j][i+1] + res[j][i-1] + res[j+1][i] + res[j-1][i])/(4)
            else:
                #print('else')
                '''variant 1'''
                res_new[j][i]=1/(2/h2**2 + 2/h1**2 - (h1*h2)**2) * (1/(h2**2)*res[j][i+1] + 1/(h2**2)*res[j][i-1] + 1/(h1**2)*res[j+1][i] + 1/(h1**2)*res[j-1][i])
                '''variant 2'''
                #res_new[j][i]=((h1**2)*(h2**2)*res[j][i] + (h2**2)*res[j][i+1] + (h2**2)*res[j][i-1] + (h1**2)*res[j+1][i] + (h1**2)*res[j-1][i])/(2*h2**2 + 2*h1**2 )
    
    return res_new

def norm(res):
    #print(res[5])
    Y,X=res.shape
    tmp=[0 for i in range(Y)]
    for j in range(Y):
            tmp[j]=max(res[j],key=abs) 
    print('\n',abs(max(tmp,key=abs)))
    return abs(max(tmp,key=abs))
```

```python
import copy
st_X=0;end_X=round(math.pi/2,6)#;print(end_X)
st_Y=0;end_Y=1#;print(end_Y)

N=11
eps=0.001

x,y = np.linspace(st_X,end_X,N),np.linspace(st_Y,end_Y,N)
print('x:',x,'\ny:',y,'\n')
#print(x,math.pi/4)
h_x=x[1]-x[0];print(h_x)
h_y=y[1]-y[0];print(h_y)


U_xy=np.zeros((len(y),len(x)))
'''заполнение сетки начальными условиями по U(0,y) & U(pi/2,y)'''
for j in range(len(y)):
    U_xy[j][0]=0       

for j in range(len(y)):  
    U_xy[j][-1]=-math.sin(y[j])
    
for i in range(len(U_xy)):
    print(U_xy[i])
    

'''вычисление нулевой сетки -  линейная интерполяция'''
res_old=linear_interpolation(x,y,U_xy,h=h_x).copy()

print('after lin interpol1\n')
for i in range(len(U_xy)):
    print(U_xy[i])


'''заполнение сетки начальными условиями попосле определения внутренностей по U(x,0) & U(x,1)'''
# тут, если делать от 0->N то нулевое условие которые мы ранее описывали в U(pi/2,y=0)=-sin(0)=0 не выполняется!!!
for i in range(1,len(x)-1):
    #U_xy[0][i]=U_xy[1][i]-h*math.sin(x[i])
    res_old[0][i]=res_old[1][i]-h_y*math.sin(x[i])

for i in range(1,len(x)-1):
    res_old[-1][i]=res_old[-2][i]/(1-h_y)

print('after lin interpol & least border conditions\n')
for i in range(len(res_old)):
    print(res_old[i])


'''заполнение внутренностей сетки по Методу Либмана'''
res_new=Liebmann_iter(res_old,h1=h_x,h2=h_y) # res_new=  вычисление первой итерация - по формуле
for i in range(len(res_new)):
    print(res_new[i],'\n')

k=1
#print('n',norm(res_new-res_old))

while norm(res_new-res_old)>eps and k<=50:
    
    res_old=res_new.copy()
    #перезаполняем краевые
    for i in range(1,len(x)-1):
        #U_xy[0][i]=U_xy[1][i]-h*math.sin(x[i])
        res_old[0][i]=res_old[1][i]-h_y*math.sin(x[i])
    for i in range(1,len(x)-1):
        res_old[-1][i]=res_old[-2][i]/(1-h_y)
    
    res_new=Liebmann_iter(res_old,h1=h_x,h2=h_y) # res_new=вычисление второй итерации - по формуле
    k+=1
    #print(norm(res_new-res_old),eps,k)# k, norm(res_new-res_old)
    print(k)

print('\nfinal table\n')
for i in range(len(res_new)):
    print(res_new[i])
    
#print(res_new[-1][-1])
#print(res_new[0][-1])
```
