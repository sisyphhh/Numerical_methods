```python
import numpy as np
from matplotlib import pylab as plt
from matplotlib import rcParams
import copy
%matplotlib inline
```

```python
def funct(x,y):
    F_xy=round(-y**2 + y/x,6)
    return F_xy
```

```python
def KKKK(x,y,h):
    
    K1=round(h*funct(x,y),6)
    
    K2=round(h*funct(x+(1/2)*h,y+(1/2)*K1),6)
    
    K3=round(h*funct(x+(1/2)*h,y+(1/2)*K2),6)
    
    K4=round(h*funct(x+h,y+K3),6)
    
    K_sum=round(K1+2*K2+2*K3+K4,6)
    
    return K_sum 
```

```python
def KKK(x,y,h):
    
    K1=round(h*funct(x,y),6)
    
    K2=round(h*funct(x+(1/3)*h,y+(1/3)*K1),6)
    
    K3=round(h*funct(x+(2/3)*h,y+(2/3)*K2),6)
    
    K_sum=round(K1+3*K3,6)
    
    return K_sum 
```

```python
def calc_function(x_,y_,hhh):
    i=1
    while x_[i-1]>0.4:
        dop=KKKK(x_[i-1],y_[i-1],hhh)

        y_.append(round(y_[i-1]-(1/6)*dop,7))#почему должны вычитать доп???
        x_.append(round(x_[i-1]-hhh,7))

        print('x_i=',x_[i],'y_i=',y_[i]) #,'F_xy=',F_xy,'h*F_xy=',h_F_xy)
        i+=1

        x_.reverse()
        y_.reverse()
        print('\n',x_)

    while x_[i-1]<2:
        dop=KKKK(x_[i-1],y_[i-1],hhh)

        y_.append(round(y_[i-1]+(1/6)*dop,7))
        x_.append(round(x_[i-1]+hhh,7))

        print('x_i=',x_[i],'y_i=',y_[i]) #,'F_xy=',F_xy,'h*F_xy=',h_F_xy)
        i+=1
        
    return
```

```python
#y`=-y^2 + y/x ,,,,x [1,2]
#fig, ax = plt.subplots(figsize=(9,12))

rcParams['figure.dpi']=150

y_0=2/3
x_0=1   # x [1,2]
h1=0.1
h2=0.01
h3=0.2


y_i=[]
x_i=[]
y_i.append(y_0)
x_i.append(x_0)

y_i2=[]
x_i2=[]
y_i2.append(y_0)
x_i2.append(x_0)

y_i3=[]
x_i3=[]
y_i3.append(y_0)
x_i3.append(x_0)

difference1_chisl=[]
difference2_chisl=[]
difference1_znam=[]
difference2_znam=[]

i=1

#реализовать ход назад и отобразаить график на интервале [0.4;3]
#те по факту на нужно реазиовать отрицательных ход до 0.4 по иск (и слелать реверс графика)
#потом от начальной узловой точки топать до 3

while x_i[i-1]>0.4:
    dop=KKKK(x_i[i-1],y_i[i-1],h1)
    
    y_i.append(round(y_i[i-1]-(1/6)*dop,7))#почему должны вычитать доп???
    x_i.append(round(x_i[i-1]-h1,7))
    
    print('x_i=',x_i[i],'y_i=',y_i[i]) #,'F_xy=',F_xy,'h*F_xy=',h_F_xy)
    i+=1
    
x_i.reverse()
y_i.reverse()
print('\n',x_i)

while x_i[i-1]<2:
    dop=KKKK(x_i[i-1],y_i[i-1],h1)
    
    y_i.append(round(y_i[i-1]+(1/6)*dop,7))
    x_i.append(round(x_i[i-1]+h1,7))
    
    print('x_i=',x_i[i],'y_i=',y_i[i]) #,'F_xy=',F_xy,'h*F_xy=',h_F_xy)
    i+=1

difference1_znam=copy.deepcopy(y_i)
plt.plot(x_i,y_i,'--',color='black',label='h1=0,1')
print('\n')
############################################################################################################
############################################################################################################
i=1
#h=0.2
while x_i3[i-1]>0.4:
    dop=KKKK(x_i3[i-1],y_i3[i-1],h3)
    
    y_i3.append(round(y_i3[i-1]-(1/6)*dop,7))
    x_i3.append(round(x_i3[i-1]-h3,7))
    
    print('x_i3=',x_i3[i],'y_i3=',y_i3[i]) #,'F_xy=',F_xy,'h*F_xy=',h_F_xy)
    i+=1
    
x_i3.reverse()
y_i3.reverse()
print('\n',x_i3)

while x_i3[i-1]<2:
    dop=KKKK(x_i3[i-1],y_i3[i-1],h3)
    
    y_i3.append(round(y_i3[i-1]+(1/6)*dop,7))
    x_i3.append(round(x_i3[i-1]+h3,7))
    
    print('x_i3=',x_i3[i],'y_i3=',y_i3[i]) #,'F_xy=',F_xy,'h*F_xy=',h_F_xy)
    i+=1
    
plt.plot(x_i3,y_i3,'--',color='green',label='h3=0,2')
print('\n')
############################################################################################################
############################################################################################################
i=1
difference2_chisl.append(y_i2[i-1])
while x_i2[i-1]>0.4:
    dop=KKKK(x_i2[i-1],y_i2[i-1],h2)
    
    y_i2.append(round(y_i2[i-1]-(1/6)*dop,7))
    x_i2.append(round(x_i2[i-1]-h2,7))
    if (i%10==0):
        print('x_i2=',x_i2[i],'y_i2=',y_i2[i])#,'F_xy=',F_xy,'h*F_xy=',h_F_xy)
        difference2_chisl.append(y_i2[i])
    i+=1

difference2_chisl.reverse()
x_i2.reverse()
y_i2.reverse()
print('\n',x_i2)

while x_i2[i-1]<2:
    dop=KKKK(x_i2[i-1],y_i2[i-1],h2)
    
    y_i2.append(round(y_i2[i-1]+(1/6)*dop,7))
    x_i2.append(round(x_i2[i-1]+h2,7))
    if (i%10==0):
        print('x_i2=',x_i2[i],'y_i2=',y_i2[i])#,'F_xy=',F_xy,'h*F_xy=',h_F_xy)
        difference2_chisl.append(y_i2[i])
    i+=1
    
#print('x_i2=',x_i2[-1],'y_i2=',y_i2[-1])#,'F_xy=',F_xy,'h*F_xy=',h_F_xy)    
    
plt.plot(x_i2,y_i2,'--',color='red',label='h2=0,01')

plt.title('Runge–Kutta method')
#plt.ylabel('Wi')
#plt.xlabel('Xi')
plt.legend()

plt.grid(True)
plt.show()
```
