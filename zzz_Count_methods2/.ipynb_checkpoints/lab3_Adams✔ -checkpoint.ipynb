```python
import numpy as np
from matplotlib import pylab as plt
from matplotlib import rcParams
%matplotlib inline
```

```python
def funct(x,y):
    F_xy=round((y*np.log(y))/x,6)
    return F_xy
```

```python
def KKKK(x,y,h):
    
    K1=round(h*funct(x,y),6)
    
    K2=round(h*funct(x+(1/2)*h,y+(1/2)*K1),6)
    
    K3=round(h*funct(x+(1/2)*h,y+(1/2)*K2),6)
    
    K4=round(h*funct(x+h,y+K3),6)
    
    K_sum=round(K1+2*K2+2*K3+K4,6)
    
    return K_sum 
```

```python
#y`=-y^2 + y/x ,,,,x [1,2]
fig, ax = plt.subplots(figsize=(16,12))

#rcParams['figure.dpi']=150

y_0=np.e
x_0=1   # x [1,2]
h1=0.1
h2=0.01


y_i=[]
x_i=[]
y_i.append(y_0)
x_i.append(x_0)

y_i2=[]
x_i2=[]
y_i2.append(y_0)
x_i2.append(x_0)


i=1
print('x_i=',x_i[i-1],'y_i=',y_i[i-1])
while x_i[i-1]<2:
    
    if i > 4:
        dop=55*funct(x_i[i-1],y_i[i-1])-59*funct(x_i[i-2],y_i[i-2])+37*funct(x_i[i-3],y_i[i-3])-9*funct(x_i[i-4],y_i[i-4])
    
        y_i.append(round(y_i[i-1]+(h1/24)*dop,7))
        x_i.append(round(x_i[i-1]+h1,7))
        print('x_i=',x_i[i],'y_i=',y_i[i]) #,'F_xy=',F_xy,'h*F_xy=',h_F_xy)
    
    else:
        dop=KKKK(x_i[i-1],y_i[i-1],h1)
    
        y_i.append(round(y_i[i-1]+(1/6)*dop,7))
        x_i.append(round(x_i[i-1]+h1,7))
        print('x_i=',x_i[i],'y_i=',y_i[i]) #,'F_xy=',F_xy,'h*F_xy=',h_F_xy)
    i+=1
        
            
plt.plot(x_i,y_i,'--',color='black',label='h1=0,1')
print('\n')


i=1
print('x_i2=',x_i2[i-1],'y_i2=',y_i2[i])
while x_i2[i-1]<2:

    if i > 4:
        dop=55*funct(x_i2[i-1],y_i2[i-1])-59*funct(x_i2[i-2],y_i2[i-2])+37*funct(x_i2[i-3],y_i2[i-3])-9*funct(x_i2[i-4],y_i2[i-4])
    
        y_i2.append(round(y_i2[i-1]+(h2/24)*dop,7))
        x_i2.append(round(x_i2[i-1]+h2,7))
        if (i%10==0):
            print('x_i2=',x_i2[i],'y_i2=',y_i2[i]) #,'F_xy=',F_xy,'h*F_xy=',h_F_xy)
        
    else:
        dop=KKKK(x_i2[i-1],y_i2[i-1],h2)
    
        y_i2.append(round(y_i2[i-1]+(1/6)*dop,7))
        x_i2.append(round(x_i2[i-1]+h2,7))
        #print('x_i2=',x_i2[i],'y_i2=',y_i2[i])#,'F_xy=',F_xy,'h*F_xy=',h_F_xy)
        
    i+=1
        
                
#print('x_i2=',x_i2[-1],'y_i2=',y_i2[-1])#,'F_xy=',F_xy,'h*F_xy=',h_F_xy)    
    
plt.plot(x_i2,y_i2,'--',color='red',label='h2=0,01')

plt.yticks(np.arange(min(y_i2), max(y_i2), 0.5))
plt.title('Adams(with Rungeâ€“Kutt) method')
#plt.ylabel('Wi')
#plt.xlabel('Xi')
plt.legend()

plt.grid(True)
plt.show()
```
