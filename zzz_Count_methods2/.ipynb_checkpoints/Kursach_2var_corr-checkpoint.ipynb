### Решить задачу коши для системы:
    y1 = x/(1 + x**2 + y2**2)**(1/2)
    y2 = y1/(1 + x**2 + y1**2)**(1/2)
На отрезке [a,b]=[-1,1] с шагом h=0.1;

    у1(а)=0.2 
    у2(а)=0.0

По таблице значений функции у1(х) построить многочлен второй степени наилучшего приближения по критерию наименьших квадратов.

```python
import numpy as np
from matplotlib import pylab as plt
from matplotlib import rcParams
%matplotlib inline
```

```python
#import math 
#fig, ax = plt.subplots(figsize=(10,10))
rcParams['figure.dpi']=150

def f1(x,y1,y2): 
    return x/(1 + x**2 + y2**2)**0.5

def f2(x,y1,y2):  
    return y1/(1 + x**2 + y1**2)**0.5

h1=0.1 
start=-1 
end=1
y1_0=0.2
y2_0=0
#----------------------------------------------------------------------
#----------------------------------------------------------------------
n1=int((end-start)/h1+1) 

x1=[round(start+i*h1,2) for i in range(n1)] 
print(x1)
y1=[0 for i in range(n1)] 
y2=[0 for i in range(n1)] 

y1[0]=y1_0 
y2[0]=y2_0


def lab4(x,y1,y2,step,mode=1): 
    n=len(x) 
    print(n) 
    s=0 
    e=n-1 
    if mode==-1: 
        s=-1 
        e=-n 
    for i in range(s,e,mode): 
        K1=[ step*f1(x[i], y1[i], y2[i]) , step*f2(x[i], y1[i], y2[i]) ] 
        K2=[ step*f1(x[i]+0.5*step, y1[i]+0.5*K1[0], y2[i]+0.5*K1[1]), step*f2(x[i]+0.5*step, y1[i]+0.5*K1[0], y2[i]+0.5*K1[1]) ] 
        K3=[ step*f1(x[i]+0.5*step, y1[i]+0.5*K2[0], y2[i]+0.5*K2[1]), step*f2(x[i]+0.5*step, y1[i]+0.5*K2[0], y2[i]+0.5*K2[1]) ] 
        K4=[ step*f1(x[i]+step, y1[i]+K3[0], y2[i]+K3[1] ), step*f2(x[i]+step, y1[i]+K3[0], y2[i]+K3[1] )] 
        delta=[1/6*( K1[0]+2*K2[0]+2*K3[0]+K4[0]),1/6*( K1[1]+2*K2[1]+2*K3[1]+K4[1])] 
        y1[i+mode]=y1[i]+delta[0] 
        y2[i+mode]=y2[i]+delta[1] 
        print('\nx=',x[i],'\ty1=',round(y1[i],6),'\ty2=',round(y2[i],6))
        #if i==0: 
            # print(y) 
    print('\nx=',x[-1],'\ty1=',round(y1[-1],6),'\ty2=',round(y2[-1],6)) 
    #print(K_1,K_2,K_3,K_4,delta,'\n') 
    #print(y) 
    return [y1,y2]
    #return y1

yR1=lab4(x1,y1,y2,h1)

import matplotlib.pyplot as plt 

fig = plt.figure(figsize=(20,10)) 
plt.plot(x1,yR1[0],'-',color='red', linewidth=2,label='y1') 
plt.scatter(x1,yR1[1],color='black', linewidth=0.5,label='y2')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#here MNK method
import numpy as np

def MNK (x, y, F_x, NxN=1):
    N=NxN+1
    X_matrix=np.zeros((N,N))
    Y_matrix=np.zeros(N)
    A_i_matrix=np.zeros(N)
    
    F_ans=np.zeros(len(F_x))
    
    #print('degree of polinom: ',NxN)
    for i in range(N):
        kk_i=NxN-i
        for j in range(N):
            #kk=N-i-j #степень в миноре икса
            kk_j=NxN-j
            for n in range(len(x)):
                X_matrix[i][j]+=(x[n]**kk_i)*(x[n]**kk_j)
    
    for i in range(N):
        kk_i=NxN-i
        for n in range(len(y)):
                Y_matrix[i]+=(y[n])*(x[n]**kk_i)
                
    print(X_matrix,'*(a)=',Y_matrix)                
    A_i_matrix = np.linalg.solve(X_matrix, Y_matrix)
    print('a_i koefs: a11*x^n+...+a1n*x^0\n',A_i_matrix,'\n\n')    
    
    for n in range(len(F_ans)):
        for i in range(len(A_i_matrix)):
            kk_i=NxN-i
            F_ans[n]+=A_i_matrix[i]*(F_x[n]**kk_i)
            
    return F_ans
```

```python
#fig, ax = plt.subplots(figsize=(10,10))
rcParams['figure.dpi']=150


num_of_dots=20
n1=int( num_of_dots + 1 ) 
f_x_m=[round(start+i*(end-start)/num_of_dots,2) for i in range(n1)] 

x1,yR1[0]
#f_x_m=np.linspace(x1[0],x1[-1],num_of_dots)
f_y_m=MNK(x1,yR1[0],f_x_m,2)#y1,degree=2

print(f_x_m)

fig = plt.figure(figsize=(10,6))#20 12

plt.plot(f_x_m,f_y_m,'--',color='blue', linewidth=2,label='MNk') 
plt.scatter(f_x_m,f_y_m,s=30,color='blue', linewidth=0.5)

plt.plot(x1,yR1[0],'-',color='red', linewidth=2,label='Runge') 
plt.scatter(x1,yR1[0],s=20,color='red', linewidth=0.5)

plt.legend()
plt.grid(True)
plt.show()


with open ('Kursach.txt','w') as outfile:
    outfile.write('x1_i'+'    '+'y_rung'+'    '+' y_mnk'+'\n')

for i in range(len(x1)):
    print('x1_i=',x1[i],'\ty_rung=',round(yR1[0][i],6),'\ty_mnk=',round(f_y_m[i],6))
    
    with open ('Kursach.txt','a') as outfile:
            outfile.write(str(x1[i])+'    '+str(round(yR1[0][i],6))+'    '+str(round(f_y_m[i],6))+'\n')
    
```
