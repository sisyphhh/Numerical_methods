```python
import numpy as np
from matplotlib import pylab as plt
from matplotlib import rcParams
%matplotlib inline
#import math
#eps=0.01
```

```python
#y_'=(-y+x*sin(x))/x [1,2]
y_0=0.3011687 #y(1)
x_0=1
h1=0.1
h2=0.01

print(np.sin(1))

# Массив значений углов, заданных в градусах:
# x = np.array([0, 30, 45, 60, 90])*np.pi/180
```

```python
#1
rcParams['figure.dpi']=150

y_i=[]
x_i=[]
y_i.append(y_0)
x_i.append(x_0)


y_i2=[]
x_i2=[]
y_i2.append(y_0)
x_i2.append(x_0)

i=1

while x_i[i-1]<2 :
    F_xy=round((-y_i[i-1]+x_i[i-1]*np.sin(x_i[i-1]))/x_i[i-1],7)
    h_F_xy=round(h1*F_xy,7)
    y_i.append(round(y_i[i-1]+h_F_xy,7))
    x_i.append(round(x_i[i-1]+h1,7))
    
    print('x_i=',x_i[i],'y_i=',y_i[i]) #,'F_xy=',F_xy,'h*F_xy=',h_F_xy)
    i+=1
    
plt.plot(x_i,y_i,'--',color='black')
print('\n')
i=1
while x_i2[i-1]<2:
    F_xy=round((-y_i2[i-1]+x_i2[i-1]*np.sin(x_i2[i-1]))/x_i2[i-1],7)
    h_F_xy=round(h2*F_xy,7)
    y_i2.append(round(y_i2[i-1]+h_F_xy,7))
    x_i2.append(round(x_i2[i-1]+h2,7))
    #if (i%49==0):
    #print('x_i2=',x_i2[i],'y_i2=',y_i2[i])#,'F_xy=',F_xy,'h*F_xy=',h_F_xy)
    i+=1
    
print('x_i2=',x_i2[-1],'y_i2=',y_i2[-1])#,'F_xy=',F_xy,'h*F_xy=',h_F_xy)    
    
plt.plot(x_i2,y_i2,'--',color='red')

plt.grid(True)
plt.show()
    
```
