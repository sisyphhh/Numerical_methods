```python
import numpy as np
from matplotlib import pylab as plt
from matplotlib import rcParams
import copy
%matplotlib inline
```

```python
def funct(x,y):
    F_xy=round(-y**2 + y/x,6)
    return F_xy
```

```python
def KKKK(x,y,h):
    
    K1=round(h*funct(x,y),6)
    
    K2=round(h*funct(x+(1/2)*h,y+(1/2)*K1),6)
    
    K3=round(h*funct(x+(1/2)*h,y+(1/2)*K2),6)
    
    K4=round(h*funct(x+h,y+K3),6)
    
    K_sum=round(K1+2*K2+2*K3+K4,6)
    
    return K_sum 
```

```python
def KKK(x,y,h):
    
    K1=round(h*funct(x,y),6)
    
    K2=round(h*funct(x+(1/3)*h,y+(1/3)*K1),6)
    
    K3=round(h*funct(x+(2/3)*h,y+(2/3)*K2),6)
    
    K_sum=round(K1+3*K3,6)
    
    return K_sum 
```

```python
#KKKK
def calc_function(x_,y_,hhh):
    i=1
    while x_[i-1]>0.4:
        dop=KKKK(x_[i-1],y_[i-1],hhh)

        y_.append(round(y_[i-1]-(1/6)*dop,7))#почему должны вычитать доп???
        x_.append(round(x_[i-1]-hhh,7))
        i+=1

    x_.reverse()
    y_.reverse()

    while x_[i-1]<2:
        dop=KKKK(x_[i-1],y_[i-1],hhh)

        y_.append(round(y_[i-1]+(1/6)*dop,7))
        x_.append(round(x_[i-1]+hhh,7))
        i+=1
     
    print('\n',x_)
    for i in range(len(x_)):
        
        if hhh==0.01:
            if (i%10==0):
                print('x_i=',x_[i],'y_i=',y_[i])#,'F_xy=',F_xy,'h*F_xy=',h_F_xy)
            else:
                continue
        else:
            print('x_i=',x_[i],'y_i=',y_[i]) #,'F_xy=',F_xy,'h*F_xy=',h_F_xy)

    answer_spisok_x_y=[]
    answer_spisok_x_y.append(x_)
    answer_spisok_x_y.append(y_)
    
    return answer_spisok_x_y
```
